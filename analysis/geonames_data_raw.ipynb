```python
import pandas as pd

from pathlib import Path
from zipfile import ZipFile
```

```python
# Special GeoNames pseudo-language / metadata tags.
# For name-generation / transduction, these are not ordinary language-name rows.
SPECIAL_ISOLANGUAGE = {
    "link",   # website link, mostly wikipedia
    "wkdt",   # wikidata id
    "post",   # postal code
    "iata",   # airport code
    "icao",   # airport code
    "faac",   # airport code
}

PROJECT_ROOT = Path.cwd().parent
RAW_DIR = PROJECT_ROOT / "data" / "raw" / "geonames"

GEONEAME_COLUMNS = [
        "geonameid",
        "name",
        "asciiname",
        "alternatenames",
        "latitude",
        "longitude",
        "feature_class",
        "feature_code",
        "country_code",
        "cc2",
        "admin1_code",
        "admin2_code",
        "admin3_code",
        "admin4_code",
        "population",
        "elevation",
        "dem",
        "timezone",
        "modification_date",
    ]

ALT_COLUMNS = [
        "alternate_name_id",
        "geonameid",
        "isolanguage",
        "alternate_name",
        "is_preferred_name",
        "is_short_name",
        "is_colloquial",
        "is_historic",
        "from_date",
        "to_date",
    ]
```

```python
df_geonames = pd.read_csv(
        RAW_DIR / "allCountries.zip",
        sep="\t",
        header=None,
        names=GEONEAME_COLUMNS,
        nrows=5000,
        compression="infer",
        encoding="utf-8",
    )
```

```python
with ZipFile(RAW_DIR / "alternateNamesV2.zip") as zf:
    with zf.open("alternateNamesV2.txt") as f:
        df_alternate_names = pd.read_csv(
            f,
            sep="\t",
            header=None,
            names=ALT_COLUMNS,
            dtype=str,
            keep_default_na=False,
            na_filter=False,
            nrows=10000,
        )
```

```python
with ZipFile(RAW_DIR / "alternateNamesV2.zip") as zf:
        with zf.open("iso-languagecodes.txt") as f:
            df_lang = pd.read_csv(
                f,
                sep="\t",
                encoding="utf-8",
                dtype=str,
            )
```
